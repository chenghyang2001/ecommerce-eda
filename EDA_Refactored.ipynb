# 電商資料探索性分析 (EDA)
模組化重構版本 — 使用 `data_loader` + `business_metrics` 模組

```python
# === 分析參數設定 ===
ANALYSIS_YEAR = 2024
COMPARISON_YEAR = 2023
DATA_PATH = 'ecommerce_data/orders.csv'
```

```python
# === 匯入模組 ===
from data_loader import EcommerceDataLoader
from business_metrics import BusinessMetricsCalculator, MetricsVisualizer
```

```python
# === 載入資料 ===
loader = EcommerceDataLoader(DATA_PATH)
df = loader.load()
df_analysis = loader.filter_by_year(ANALYSIS_YEAR)
print(f'全部資料: {len(df)} 筆')
print(f'{ANALYSIS_YEAR} 年資料: {len(df_analysis)} 筆')
print(f'日期範圍: {loader.get_date_range()}')
```

```python
# === 計算業務指標 ===
calc = BusinessMetricsCalculator(df_analysis)
rev = calc.revenue_metrics()
print(f"總營收: ${rev['total_revenue']:,.2f}")
print(f"總訂單: {rev['total_orders']}")
print(f"平均訂單金額: ${rev['avg_order_value']:,.2f}")
print(f"最新月成長率: {rev['latest_growth']:.1f}%")
```

```python
# === 營收趨勢圖 ===
viz = MetricsVisualizer()
viz.revenue_trend(rev['monthly_revenue']).show()
```

```python
# === 產品類別表現 ===
products = calc.product_metrics()
print(products.to_string(index=False))
viz.category_performance(products).show()
```

```python
# === 各州營收 ===
geo = calc.geo_metrics()
viz.state_revenue(geo).show()
```

```python
# === 客戶滿意度 ===
sat = calc.satisfaction_metrics()
print(f"平均滿意度: {sat['mean']:.2f}")
print(f"中位數: {sat['median']:.0f}")
viz.satisfaction_distribution(df_analysis).show()
```

```python
# === 配送指標 ===
delivery = calc.delivery_metrics()
print(f"平均配送天數: {delivery['avg_days']:.1f}")
print(f"超過 7 天比例: {delivery['over_7_days_pct']:.1f}%")
viz.delivery_vs_satisfaction(delivery['by_satisfaction']).show()
```

```python
# === 年度比較 ===
for year in [COMPARISON_YEAR, ANALYSIS_YEAR]:
    df_year = loader.filter_by_year(year)
    c = BusinessMetricsCalculator(df_year)
    r = c.revenue_metrics()
    s = c.satisfaction_metrics()
    print(f'\n--- {year} ---')
    print(f"總訂單: {r['total_orders']}")
    print(f"總營收: ${r['total_revenue']:,.2f}")
    print(f"平均訂單金額: ${r['avg_order_value']:,.2f}")
    print(f"平均滿意度: {s['mean']:.2f}")
```
